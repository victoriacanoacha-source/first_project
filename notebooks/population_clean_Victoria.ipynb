```python
import pandas as pd

df = pd.read_csv("country_pop.csv", sep=",", skiprows=3)
df.columns

df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(" ", "_"))

df = df.rename(columns={"country_name":"country", "country_code": "code"})
df
```

```python
pop_long = (
    df.melt(
        id_vars=["country", "code"],
        var_name="year",
        value_name="population"))


pop_long["year"] = pd.to_numeric(pop_long["year"], errors="coerce")
pop_long = pop_long.dropna(subset=["year", "population"])

pop_long = pop_long[(pop_long["year"] >= 1990)]
pop_long["year"] = pop_long["year"].astype(int)
pop_long
```

```python
pop_long.to_csv("pop_dataset.csv", index=False)
```
