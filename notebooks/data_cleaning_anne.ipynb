```python
import pandas as pd
import numpy as np

# Load the dataset Religions
src = r'C:\Users\Utilisateur\Desktop\IRONHACK_DA\COURSES\week-4\first_project\data\raw/ThrowbackDataThursday 201912 - Religion.csv'
df = pd.read_csv(src)

# Drop duplicates and na, adding a column name for index
df.drop_duplicates(inplace=True)
df.dropna()

# Filter from 1990 to 2022
df = df[(df.Year >=1990) & (df.Year <=2022)]

# Keep state, year, total and percentage per religion
col_keep = [col for col in df.columns if col.endswith("Total (Population)") or col.endswith("Total (Percent)") or col == "Year" or col == "StateNme"]
df = df[col_keep]

# Rename columns
df.columns = [col.replace("- Total (Percent)", "_pct").replace(" - Total (Population)", "_pop")
    .replace("StateNme","Country") for col in df.columns]

# Remove rows with total percentages over 1
df = df[df['Total Religious _pct'] <= 1]

# df.to_csv(r"C:\Users\Utilisateur\Desktop\IRONHACK_DA\COURSES\week-4\first_project\data\raw/check.csv")
```

```python
# Load the dataset of country referencing
src1 = r'C:\Users\Utilisateur\Desktop\IRONHACK_DA\COURSES\week-4\first_project\data\raw\JME_Regional-Classifications.csv'
ref_country = pd.read_csv(src1, sep = ';')

# Rename the columns we want 
ref_country.columns = [col.replace("UN Region", "Region").replace("UN Sub Region", "SubRegion") for col in ref_country.columns]

# Get the unique values of country
countries = pd.DataFrame(df.Country.unique())
countries.columns = ["Country"]

# Get the countries that do no match with the official list
missing_countries = countries[~countries['Country'].isin(ref_country['Country'])]


# missing_countries.to_csv(r"C:\Users\Utilisateur\Desktop\IRONHACK_DA\COURSES\week-4\first_project\data\raw/missing_countries.csv")
```

```python
# Replace the countries names accordingly

country_mapping = {
    "Antigua & Barbuda": "Antigua and Barbuda",
    "Bolivia": "Bolivia (Plurinational State of)",
    "Brunei": "Brunei Darussalam",
    "Cape Verde": "Cabo Verde",
    "Czech Republic": "Czechia",
    "Czechoslovakia": "Slovakia",
    "East Timor": "Timor-Leste",
    "Federated States of Micronesia": "Micronesia (Federated States of)",
    "German Federal Republic": "Germany",
    "Iran": "Iran (Islamic Republic of)",
    "Ivory Coast": "Côte d'Ivoire",
    "Laos": "Lao People's Democratic Republic",
    "Macedonia": "The former Yugoslav Republic of Macedonia",
    "Moldova": "Republic of Moldova",
    "North Korea": "Democratic People's Republic of Korea",
    "Russia": "Russian Federation",
    "South Korea": "Republic of Korea",
    "St. Kitts and Nevis": "Saint Kitts and Nevis",
    "St. Lucia": "Saint Lucia",
    "St. Vincent and the Grenadines": "Saint Vincent and the Grenadines",
    "Syria": "Syrian Arab Republic",
    "Tanzania": "United Republic of Tanzania",
    "United States of America": "United States",
    "Venezuela": "Venezuela (Bolivarian Republic of)",
    "Vietnam": "Viet Nam",
    "Yemen Arab Republic": "Yemen",
    "Yemen People's Republic": "Yemen"
}

df['Country'] = df['Country'].replace(country_mapping)
```

```python
# Remove the countries with no match
df = df[~ df['Country'].isin(["Kosovo", "Taiwan", "Yugoslavia"])].reset_index()
```

```python
# Join columns to get Regions and Subregions
df = df.merge(ref_country[['Country', 'Region','SubRegion']], on='Country', how='left')
```

```python
# Export to csv
df = df.reset_index(drop=True) 
df.to_csv(r"C:\Users\Utilisateur\Desktop\IRONHACK_DA\COURSES\week-4\first_project\data\raw\clean_religions.csv", index = False)
```
