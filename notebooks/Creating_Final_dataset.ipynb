```python
import pandas as pd

religion = pd.read_csv("clean_religions.csv")
gdp = pd.read_csv("gdp_data_cleaned_file.csv")
hdi = pd.read_csv("hdi_long.csv")
conflict1 = pd.read_csv("conflict_countries.csv")
regions= pd.read_csv("JME_Regional-Classifications.csv", sep = ";")
conflict2 = pd.read_csv("Conflict_Site_from_2005.csv", sep = ";")
```

```python
religion = religion.rename(columns={"Country":"country","Year":"year"})
conflict1 = conflict1.rename(columns={"Country":"country","Year":"year"})
conflict2 = conflict2.rename(columns={"Country":"country","Year":"year"})
regions.columns = (
        regions.columns
        .str.strip()
        .str.lower()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(" ", "_"))
regions
```

```python
conflict1["conflict_present"] = 1
conflict2["conflict_present"] = 1
```

```python
conflict = pd.concat([conflict1, conflict2], axis=0, join="outer")
conflict.drop_duplicates(inplace=True)
conflict 
```

```python
merged = religion.merge(gdp, on=["country","year"],how="left")
```

```python
merged.tail()
```

```python
merged = merged.merge(hdi, on=["country","year"],how="left")
merged = merged.merge(conflict, on=["country","year"],how="left")

merged["country"] = merged["country"].str.title()
regions["country"] = regions["country"].str.title()

merged = merged.merge(regions, on=["country", "un_region"],how="left")
```

```python
merged["conflict_present"] = merged["conflict_present"].fillna(0)
```

```python
merged = merged.drop(['index'], axis=1)
merged.drop_duplicates(inplace=True)
```

```python
#Load the updated HDI dataset containing yearly HDI values and dropping the old hdi column to prevent duplicated before merging the datasets
#Then we merged the updated HDI dataset with the main dataset using country and year as the keys ensuring that each country–year observation receives the correct HDI value.
hdi_new = pd.read_csv("hdi_long_final.csv")
merged = merged.drop(columns=["hdi"], errors="ignore")
merged = merged.merge(hdi_new,on=["country","year"],how="left")

#Check the number of null values and identify which countries contain missing HDI values to understand where data gaps exist in the HDI dataset
#Check whether the region classifications from the original dataset and the HDI dataset are identical across all observations.
#Count how many rows contain different region classifications between the two datasets.
#Examine the rows with missing HDI values to determine whether region mismatches occur due to missing HDI data.

merged["hdi"].isna().sum()
merged[merged["hdi"].isna()].groupby("country").size()
(merged["un_region_x"] == merged["un_region_y"]).all()
(merged["un_region_x"] != merged["un_region_y"]).sum()
merged[merged["hdi"].isna()][["country","year","un_region_x","un_region_y"]].head()
```

```python
#Remove the region column originating from the HDI dataset to avoid duplicate region variables.
#Rename the remaining region column for clarity and consistency within the dataset.

merged = merged.drop(columns=["un_region_y"])
merged = merged.rename(columns={"un_region_x": "un_region"})
```

```python
#Standardised column names so they all follow the same format 
merged.columns = (merged.columns.str.strip().str.lower())
```
